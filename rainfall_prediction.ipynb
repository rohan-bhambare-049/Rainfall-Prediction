```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
df = pd.read_csv('rain_forecasting.csv')
# Conversion of Categorical to Numeric
df['RainTomorrow'] = df['RainTomorrow'].map({'Yes': 1, 'No': 0})
df['RainToday'] = df['RainToday'].map({'Yes': 1, 'No': 0})
```

```python
# Bar Chart
plt.figure(figsize=(6,4))
df['RainTomorrow'].value_counts().plot(kind='bar')
plt.xlabel('Rain Tomorrow (0=No, 1=Yes)')
plt.ylabel('Number of Days')
plt.title('Rainy vs Non Rainy Days')
plt.show()
```

```python
# Histogram
df[['MinTemp', 'MaxTemp', 'Humidity9am', 'Humidity3pm']].hist(bins=15, figsize=(10,6))
plt.suptitle('Distributions: Temperature and Humidity')
plt.show()
```

```python
# 3. Scatter Graph
plt.figure(figsize=(6,4))
sns.scatterplot(x='MaxTemp', y='Humidity3pm', hue='RainTomorrow', data=df)
plt.title('MaxTemp vs Humidity3pm (RainTomorrow Highlighted)')
plt.show()
```

```python
# Selection of Features for Prediction
features = ['MinTemp', 'MaxTemp', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'WindSpeed9am', 'WindSpeed3pm']
```

```python
# Handling Missing Values (if any)
df = df.dropna()

X = df[features]
y = df['RainTomorrow']
```

```python
# Split into Train and Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Train Decision Tree Classifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
# Predict on Test Data
y_pred = model.predict(X_test)
```

```python
# Print accuracy
print('Accuracy:', accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
```

```python
# Load data
df = pd.read_csv('rain_forecasting.csv')
df['RainTomorrow'] = df['RainTomorrow'].map({'Yes': 1, 'No': 0})
df['RainToday'] = df['RainToday'].map({'Yes': 1, 'No': 0})
df = df.dropna()

X = df[['MinTemp', 'MaxTemp', 'Humidity9am', 'Humidity3pm',
        'Pressure9am', 'Pressure3pm', 'WindSpeed9am', 'WindSpeed3pm','RainToday']]
y = df['RainTomorrow']
```

```python
# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# SMOTE for balancing classes
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=0)
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=0)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
```

```python
# Predict on Test Data
y_pred = model.predict(X_test)
```

```python
# Example: Predict Rainfall for Tommorow
new_data = pd.DataFrame({
    'MinTemp': [25],
    'MaxTemp': [35],
    'Humidity9am': [80],
    'Humidity3pm': [60],
    'Pressure9am': [1010],
    'Pressure3pm': [1007],
    'WindSpeed9am': [12],
    'WindSpeed3pm': [15],
    'Rain Today':[1]
})
rain_pred = model.predict(new_data)
print("Rain Tomorrow (1=Yes, 0=No):", rain_pred[0])
```
